```python
#set colors
import matplotlib.pyplot as plt

# ── Theme colors ───────────────────────────────────────────
GES_DARK  = '#01010b'   # black
GES_BLUE  = '#6950ec'   # purple
GES_GOLD  = '#e00e32'   # berry red
GES_TEAL  = '#a3055c'   # maroon

PALETTE = [
    '#6950ec',  # purple
    '#a3055c',  # maroon
    '#e00e32',  # berry red
    '#9e9e9e',  # gray
    '#8a7cff',  # soft purple
    '#c75a93',  # dusty pink
    '#d9d9d9',  # light gray
    '#b23a6f'   # muted berry
]

# ── Global chart settings ──────────────────────────────────
FIGSIZE = (10, 6)
TITLE_SIZE = 14
LABEL_SIZE = 11
TICK_SIZE = 10
BAR_LABEL_SIZE = 9
ROTATION = 20
GRID_ALPHA = 0.35
BAR_ALPHA = 0.9
EDGE_COLOR = '#d9d9d9'

plt.rcParams.update({
    'figure.facecolor':  GES_DARK,
    'axes.facecolor':    '#141420',
    'savefig.facecolor': GES_DARK,
    'axes.edgecolor':    '#444',
    'axes.labelcolor':   'white',
    'xtick.color':       '#ccc',
    'ytick.color':       '#ccc',
    'text.color':        'white',
    'grid.color':        '#333',
    'grid.linestyle':    '--',
    'grid.alpha':        0.5,
    'axes.spines.top':   False,
    'axes.spines.right': False,
})
```

```python
import pandas as pd

df = pd.read_csv("Data/merged_customer_data.csv")
df.head()
```

### Date setup

```python
# Convert date columns
date_cols = ["Transaction Date", "First Interaction", "Last Interaction"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# Clean numeric fields
df["Revenue"] = pd.to_numeric(df["Revenue"], errors="coerce")

# Drop bad rows for time analysis
df = df.dropna(subset=["Transaction Date"]).copy()

# Sort
df = df.sort_values("Transaction Date")

# Add time fields
df["Year"] = df["Transaction Date"].dt.year
df["Month"] = df["Transaction Date"].dt.month
df["Month Start"] = df["Transaction Date"].dt.to_period("M").dt.to_timestamp()

print("Shape:", df.shape)
```

## Looking at monthly transactions

All data is from 2024

```python
import matplotlib.dates as mdates

monthly_txns = (
    df.groupby("Month Start")
      .size()
      .reset_index(name="Transactions")
)

fig, ax = plt.subplots(figsize=FIGSIZE)
ax.plot(
    monthly_txns["Month Start"],
    monthly_txns["Transactions"],
    color=PALETTE[0],
    linewidth=2.5
)

ax.set_title("Monthly Transaction Volume", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Month", fontsize=LABEL_SIZE)
ax.set_ylabel("Number of Transactions", fontsize=LABEL_SIZE)
ax.grid(True, axis="y", alpha=GRID_ALPHA)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))

ax.set_ylim(175, 250)
plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

```python
monthly_revenue = (
    df.groupby("Month Start")["Revenue"]
      .sum()
      .reset_index()
)

fig, ax = plt.subplots(figsize=FIGSIZE)
ax.plot(
    monthly_revenue["Month Start"],
    monthly_revenue["Revenue"],
    color=PALETTE[2],
    linewidth=2.5
)

ax.set_title("Monthly Revenue Trend", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Month", fontsize=LABEL_SIZE)
ax.set_ylabel("Revenue", fontsize=LABEL_SIZE)
ax.grid(True, axis="y", alpha=GRID_ALPHA)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))

plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

Very similar distribution

## Missing identifiers time analysis

```python
# ─────────────────────────────────────────────
# 1 — Missing identifiers over time (MONTH, overlap-aware)
# ─────────────────────────────────────────────
missing_month = (
    df.groupby("Month Start")
      .apply(lambda g: pd.Series({
          "Email Missing": g["Email"].isna().mean() * 100,
          "Age & Gender Both Missing": ((g["Age Range"].isna()) & (g["Gender"].isna())).mean() * 100
      }))
      .reset_index()
)

fig, ax = plt.subplots(figsize=FIGSIZE)

ax.plot(
    missing_month["Month Start"],
    missing_month["Email Missing"],
    marker="o",
    linewidth=2.5,
    color=PALETTE[0],
    label="Email Missing"
)

ax.plot(
    missing_month["Month Start"],
    missing_month["Age & Gender Both Missing"],
    marker="o",
    linewidth=2.5,
    color=PALETTE[2],
    label="Age & Gender Both Missing"
)

ax.set_title("Missing Identifier Rates Over Time", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Month", fontsize=LABEL_SIZE)
ax.set_ylabel("% Missing", fontsize=LABEL_SIZE)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
ax.grid(True, axis="y", alpha=GRID_ALPHA)
ax.legend(frameon=False)
plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

- Age and Demographic happen exactly concurrently
- Email has some deviations

## Montly revenue by business line

```python
# ─────────────────────────────────────────────
# 2 — Monthly revenue seasonality by business line
#    (manual colors to avoid similar shades)
# ─────────────────────────────────────────────
rev_bl = (
    df.groupby(["Month Start", "Business Line"])["Revenue"]
      .sum()
      .reset_index()
)

fig, ax = plt.subplots(figsize=(12, 6))

business_lines = sorted(rev_bl["Business Line"].dropna().unique())

#change color template temporariliy
line_colors = [
    GES_BLUE,      # purple
    GES_TEAL,      # maroon
    GES_GOLD,      # berry red
    "#9e9e9e",     # gray
    "#ff6b6b",     # lighter red/coral
    "#4cc9f0",     # cyan
    "#f4d35e",     # gold/yellow
    "#b8b8ff"      # pale lavender
]

for i, bl in enumerate(business_lines):
    temp = rev_bl[rev_bl["Business Line"] == bl]
    ax.plot(
        temp["Month Start"],
        temp["Revenue"],
        marker="o",
        linewidth=2.5,
        color=line_colors[i % len(line_colors)],
        label=bl
    )

ax.set_title("Monthly Revenue Seasonality by Business Line", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Month", fontsize=LABEL_SIZE)
ax.set_ylabel("Revenue", fontsize=LABEL_SIZE)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
ax.grid(True, axis="y", alpha=GRID_ALPHA)
ax.legend(frameon=False, title="Business Line")
plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

Pretty consistent revenue throughout but Streaming spiked in March and Studio tours spiked in July

## Monthly revenue by source type

```python
# ─────────────────────────────────────────────
# 3 — Monthly revenue seasonality by source type
# ─────────────────────────────────────────────
rev_source = (
    df.groupby(["Month Start", "Source Type"])["Revenue"]
      .sum()
      .reset_index()
)

fig, ax = plt.subplots(figsize=FIGSIZE)

source_colors = {
    "1st Party": GES_BLUE,
    "3rd Party": GES_GOLD
}

for source in sorted(rev_source["Source Type"].dropna().unique()):
    temp = rev_source[rev_source["Source Type"] == source]
    ax.plot(
        temp["Month Start"],
        temp["Revenue"],
        marker="o",
        linewidth=2.7,
        color=source_colors.get(source, "#9e9e9e"),
        label=source
    )

ax.set_title("Monthly Revenue Seasonality by Source Type", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Month", fontsize=LABEL_SIZE)
ax.set_ylabel("Revenue", fontsize=LABEL_SIZE)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
ax.grid(True, axis="y", alpha=GRID_ALPHA)
ax.legend(frameon=False)
plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

1st Party drives far more revenue every month and has the bigger swings, with clear peaks around March and July and softer periods around April–June. 3rd Party stays much lower overall, but trends a little more stable with a modest lift toward September–Decembe

## Weekly trends

```python
# ─────────────────────────────────────────────
# 4 — Revenue by day of week
#    built directly from Transaction Date
# ─────────────────────────────────────────────
dow_df = df.copy()

dow_df["Transaction Date"] = pd.to_datetime(dow_df["Transaction Date"])
dow_df["Day Num"] = dow_df["Transaction Date"].dt.dayofweek
dow_df["Day Name"] = dow_df["Transaction Date"].dt.day_name()

dow_rev = (
    dow_df.groupby(["Day Num", "Day Name"])["Revenue"]
          .sum()
          .reset_index()
          .sort_values("Day Num")
)

fig, ax = plt.subplots(figsize=FIGSIZE)

bars = ax.bar(
    dow_rev["Day Name"],
    dow_rev["Revenue"],
    color=[PALETTE[i % len(PALETTE)] for i in range(len(dow_rev))],
    alpha=BAR_ALPHA,
    edgecolor=EDGE_COLOR
)

ax.set_title("Revenue by Day of Week", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Day of Week", fontsize=LABEL_SIZE)
ax.set_ylabel("Revenue", fontsize=LABEL_SIZE)
ax.grid(True, axis="y", alpha=GRID_ALPHA)
ax.bar_label(bars, padding=3, fontsize=BAR_LABEL_SIZE, fmt="%.0f", color="white")
plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

No crazy outliers, but more Wednesday and Saturday

```python
# ─────────────────────────────────────────────
# 4B — Transactions by day of week
#    built directly from Transaction Date
# ─────────────────────────────────────────────
dow_df = df.copy()

dow_df["Transaction Date"] = pd.to_datetime(dow_df["Transaction Date"])
dow_df["Day Num"] = dow_df["Transaction Date"].dt.dayofweek
dow_df["Day Name"] = dow_df["Transaction Date"].dt.day_name()

dow_txn = (
    dow_df.groupby(["Day Num", "Day Name"])
          .size()
          .reset_index(name="Transactions")
          .sort_values("Day Num")
)

fig, ax = plt.subplots(figsize=FIGSIZE)

bars = ax.bar(
    dow_txn["Day Name"],
    dow_txn["Transactions"],
    color=[PALETTE[i % len(PALETTE)] for i in range(len(dow_txn))],
    alpha=BAR_ALPHA,
    edgecolor=EDGE_COLOR
)

ax.set_title("Transactions by Day of Week", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Day of Week", fontsize=LABEL_SIZE)
ax.set_ylabel("Transactions", fontsize=LABEL_SIZE)
ax.grid(True, axis="y", alpha=GRID_ALPHA)
ax.bar_label(bars, padding=3, fontsize=BAR_LABEL_SIZE, color="white")
plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

## New customers over time

```python
# ─────────────────────────────────────────────
# 5B — New customers over time (line chart)
# ─────────────────────────────────────────────
new_cust_df = df.copy()
new_cust_df["Transaction Date"] = pd.to_datetime(new_cust_df["Transaction Date"])

first_txn = (
    new_cust_df.groupby("Customer ID")["Transaction Date"]
               .min()
               .reset_index(name="First Transaction")
)

first_txn["Month Start"] = first_txn["First Transaction"].dt.to_period("M").dt.to_timestamp()

new_customers = (
    first_txn.groupby("Month Start")
             .size()
             .reset_index(name="New Customers")
)

fig, ax = plt.subplots(figsize=FIGSIZE)

ax.plot(
    new_customers["Month Start"],
    new_customers["New Customers"],
    marker="o",
    linewidth=2.5,
    color=PALETTE[3]
)

ax.set_title("New Customers Over Time", fontsize=TITLE_SIZE, fontweight="bold")
ax.set_xlabel("Month", fontsize=LABEL_SIZE)
ax.set_ylabel("New Customers", fontsize=LABEL_SIZE)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
ax.grid(True, axis="y", alpha=GRID_ALPHA)

plt.xticks(rotation=ROTATION)
plt.tight_layout()
plt.show()
```

Elbow-like decline and hits single digits towards the second half of the year

```python
# ─────────────────────────────────────────────
# 6 — Multi-channel conversion lag
#    days from first transaction to second business line
# ─────────────────────────────────────────────
mc_df = df.copy()
mc_df["Transaction Date"] = pd.to_datetime(mc_df["Transaction Date"])

# sort so customer histories are in order
mc_df = mc_df.sort_values(["Customer ID", "Transaction Date"])

conversion_rows = []

for cust_id, group in mc_df.groupby("Customer ID"):
    seen_lines = set()
    first_touch = group["Transaction Date"].min()
    conversion_date = pd.NaT
    
    for _, row in group.iterrows():
        seen_lines.add(row["Business Line"])
        if len(seen_lines) >= 2:
            conversion_date = row["Transaction Date"]
            break
    
    conversion_rows.append({
        "Customer ID": cust_id,
        "First Touch": first_touch,
        "Conversion Date": conversion_date
    })

conversion_lag = pd.DataFrame(conversion_rows)
conversion_lag["Lag Days"] = (
    conversion_lag["Conversion Date"] - conversion_lag["First Touch"]
).dt.days

# keep only customers who actually became multi-channel
converted = conversion_lag[conversion_lag["Lag Days"].notna()].copy()

print("Converted customers:", len(converted))
print(converted["Lag Days"].describe())
```
